```python
img_list = []
#foldname = 'exp_edge_gripper/'
foldname = 'exp_edge_button/'

for i in range(30):
    num =str(i)
    file_name = num.zfill(2) 
    #print(foldname+file_name + '.tiff')
    img = cv.imread(foldname+file_name + '.tiff',cv.IMREAD_ANYCOLOR | cv.IMREAD_ANYDEPTH)

    # swap color channels to XYZ
    XYZimg = img[:,:,::-1]
    #print(rgbimg.shape)
    #cv.imshow("TIF opencv swap: rgb", XYZimg)
    #cv.waitKey(0)
    #cv.destroyWindow("TIF opencv swap: rgb")

    img_list.append(XYZimg)
    
print('loaded img:',len(img_list))
```

```python
img = img_list[8]
print(img.shape)
x = img[:,:,0]
y = img[:,:,1]
z = img[:,:,2]
idx = z > 1.0
print(idx.shape)
Xp = x[idx]
Yp = y[idx]
Zp = z[idx]
print(Zp.shape)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# plot raw data
#plt.figure()
#plt.figure(figsize=(20,10))
#ax = plt.subplot(111, projection='3d')
#ax.scatter(Xp, Yp, Zp, color='g')
```

```python
# do fit
#tmp_A = []
#tmp_b = []
#for i in range(len(Xp)):
#    tmp_A.append([Xp[i], Yp[i], 1])
#    tmp_b.append(Zp[i])
    
#print(A.T*A)
#print((A.T*A).I)
#print('I:',(A.T*A).I*(A.T*A))
 
Op = np.ones((Xp.shape[0]))
Atmp = np.vstack([Xp,Yp,Op])
print(type(Atmp))
    
b = np.matrix(Zp).T
A = np.matrix(Atmp.T)
```

```python
print(b.shape)
print(A.shape)

# print(A.T*A)
```

```python
# .I is inverse matrix
fit = (A.T * A).I * A.T * b
errors = b - A * fit
residual = np.linalg.norm(errors)

print ("solution:")
print ("%f x + %f y + %f = z" % (fit[0], fit[1], fit[2]))
print ("errors:",errors)
print ("residual:",residual)
```

```python
# plot plane
# plot raw data
#plt.figure()
plt.figure(figsize=(20,10))
ax = plt.subplot(111, projection='3d')
ax.scatter(Xp, Yp, Zp, color='g')

xlim = ax.get_xlim()
ylim = ax.get_ylim()
X,Y = np.meshgrid(np.arange(xlim[0], xlim[1]),
                  np.arange(ylim[0], ylim[1]))
Z = np.zeros(X.shape)
for r in range(X.shape[0]):
    for c in range(X.shape[1]):
        Z[r,c] = fit[0] * X[r,c] + fit[1] * Y[r,c] + fit[2]
ax.plot_wireframe(X,Y,Z, color='grey')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
# Ax + By -Cz = -D
# C is 1
# normalize vector AB-C
vecN = fit.copy()
vecN[2] = -1 # points in z- direction
lenV = np.linalg.norm(vecN)
print('length:',lenV)
if lenV > 0:
    vecN = vecN / lenV

#vecN = -vecN    
    
print(fit)
print('length',np.linalg.norm(vecN))
print(vecN)

vecABCD = np.vstack((vecN,fit[2]/lenV))
print(vecABCD)
```

```python
ResOrig = fit[0] * R[:,0] + fit[1] * R[:,1] + fit[2]
print(np.around(ResOrig,4))
ResNorm = (vecABCD[0] * R[:,0] + vecABCD[1] * R[:,1] + vecABCD[3])/(-vecABCD[2])
print(ResNorm)
```

```python
# get point on surface at X:0,Y:0 on plane [x: 0.0, y: 0.0]
P1 = np.array([0.0,0.0,0.0])
P1[2] = vecABCD[3]/(-vecABCD[2])
print('P1 in center [x,y,z]:\n',P1)

# get point on surface in X+ direction [x: 100.0, y: 0.0]
P2 = np.array([100.0,0.0,0.0])
P2[2] = (vecABCD[0] * 100.0 + vecABCD[3])/(-vecABCD[2])
print('P2 [x,y,z]:\n',P2)

#e1 = ',u - o) / np.linalg.norm(u-o) # x
#e2 = (v - o) / np.linalg.norm(v-o) # y
#e3 = np.cross(e1,e2) # z
```

```python
# union vector on surface should look in z+ direction!

tmp = np.copy(vecN) # creates array!!!
#e1 = vecN.copy() # creates matrix!!!
e3 = np.squeeze(tmp) # plane normal is Z direction
#print('e1:',e1)
e1 = (P2-P1) / np.linalg.norm(P2-P1)
#print('e2:',e2)

# both should be matrix for cross product
#print(e1.shape)
#print(e2.shape)
#print(type(e1))
#print(type(e2))
e2 = np.cross(e3,e1)

print('ex1:',e1)
print('ey2:',e2)
print('ez3:',e3)
```

```python
# set unit vectors as columns to rotation matrix
R = np.concatenate([[e1],[e2],[e3]])
R
```
